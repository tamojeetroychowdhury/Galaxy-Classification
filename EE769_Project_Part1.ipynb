As could be seen from te website, the number of images in class 5 (edge-on with boxy bulge) had only 17 images and therefore impossible to train the model on usefully. As a result, we decided to exclude these galaxies completely from our analysis of the ML model and only work with the remaining nine classes. The labels for all classes beyod 5 is hence shifted down by one so that the labels for our final dataset go as 0 to 8.

Also, our dataset has all images in order i.e. all 0's together, 1's together after that and so on. As a result if I directly pick the first 16000 images for training, it will have the entirety of first few classes and miss all images of last few. To make sure that doesn't happen, np.random.shuffle is used on the images and the labels to rearrange them randomly and ensure a proportionate presence of all labels in the final training and validationn sets.

```python
images, labels = load_galaxy10sdss()

ind = np.where(labels!=5)
images, labels = images[ind], labels[ind]

ind = np.arange(len(labels))
np.random.shuffle(ind)
images, labels = images[ind], labels[ind]
labels[np.where(labels > 5)] -= 1
labels = np.array(labels, dtype='int')

plt.imshow(images[1010])
```

Making sure that the labels are 0 to 8 and nothing else i.e. that there is no missing data.

For training the PyTorch model, we cannot pass in arrays directly and we need to create dataloaders of torch tensors. We hence do the same in the cell below, by using a class definition that returns these created dataloaders from the raw RGB image arrays.

Additionally, out of about 21000 images, we do a train-test split at 16000, which is roughly an 80% - 20% split between training and validation.

```python
class CustomDataset(Dataset):
    def __init__(self, images, labels):
        self.images = images
        self.labels = labels

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        image = self.images[idx]
        label = self.labels[idx]

        # Transpose image array to match PyTorch's tensor format (C x H x W)
        image = np.transpose(image, (2, 0, 1))

        # Convert numpy array to PyTorch tensor
        image = torch.tensor(image, dtype=torch.float)
        label = torch.tensor(label, dtype=torch.long)

        return image, label


# Split dataset into train and validation sets
train_images, val_images = images[:16000], images[16000:]
train_labels, val_labels = labels[:16000], labels[16000:]

# Define datasets and dataloaders
train_dataset = CustomDataset(train_images, train_labels)
val_dataset = CustomDataset(val_images, val_labels)

# Assuming batch size is 32
batch_size = 32

# Create dataloaders
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
```

### Creating the CNN architecture model and training it:

We first define our neural net with a variable kernel size. The original paper did not specify the kernel size used, so we contacted the authors and got their kernel size to be 3 which was the same value that we used as well henceforth.

In the training loop, we run the training with batch sizes of 32, a standard number in many such ML trainings. Cross-entropy Adam loss is used for optimization, which is again a standard rubric to use. WE subsequently print the training losses and accuracies after every 100 batches in each epoch. Since there are 16000 with batch size of 32, number of batches is 500.

After every epoch we also print the classification accuracy on the validation dataset, using net.eval and net.train mode functions.

```python
m=256
class Net(nn.Module):
    def __init__(self, k_size=3):
        super(Net, self).__init__()
        self.conv0 = nn.Conv2d(3, 32, kernel_size=k_size)
        self.conv1 = nn.Conv2d(32, 64, kernel_size=k_size)
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(64, 64, kernel_size=k_size)
        #self.conv3 = nn.Conv2d(20, 30, kernel_size=k_size)
        self.fc1 = nn.Linear(m, 64)
        self.fc2 = nn.Linear(64, 32)
        self.fc3 = nn.Linear(32, 9)

    def forward(self, x):
        x = F.relu(self.conv0(x))
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, m)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        #x = F.softmax(x, dim=1)
        return x

# Example usage:
net = Net(k_size=3)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(net.parameters(), lr=0.001)

# Number of epochs
epochs = 10
curr_val_acc = prev_val_acc = 0

# Lists to store loss and accuracy for plotting
train_losses = []
train_accuracies = []
val_accuracies = []

# Training loop
for epoch in range(epochs):
    running_loss = 0.0
    correct = 0
    total = 0
    for i, data in enumerate(train_loader, 0):
        inputs, labels = data
        optimizer.zero_grad()

        # Forward pass
        outputs = net(inputs)

        # Compute loss
        loss = criterion(outputs, labels)

        # Backward pass and optimize
        loss.backward()
        optimizer.step()

        # Compute training accuracy
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        # Print training statistics
        running_loss += loss.item()
        if i % 100 == 99:    # Print every 100 mini-batches
            print('[%d, %5d] train loss: %.3f, train accuracy: %.2f %%' %
                  (epoch + 1, i + 1, running_loss / 100, 100 * correct / total))
            train_losses.append(running_loss / 100)
            train_accuracies.append(100 * correct / total)
            running_loss = 0.0
            correct = 0
            total = 0

    # Compute validation accuracy
    net.eval()
    with torch.no_grad():
        correct_val = 0
        total_val = 0
        for data in val_loader:
            inputs_val, labels_val = data
            outputs_val = net(inputs_val)
            _, predicted_val = torch.max(outputs_val, 1)
            total_val += labels_val.size(0)
            correct_val += (predicted_val == labels_val).sum().item()
        val_accuracy = 100 * correct_val / total_val
        prev_val_acc, curr_val_acc = curr_val_acc, val_accuracy
        val_accuracies.append(val_accuracy)
        print('[%d] validation accuracy: %.2f %%' % (epoch + 1, val_accuracy))

    if curr_val_acc - prev_val_acc <= 0.5:
        break
    net.train()

```

Now we plot the training and validation losses and accuracies w.r.t. each epoch.

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
plt.plot(np.arange(1, len(train_losses)+1)/5 , train_losses, marker='o')
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
plt.title('Training Loss vs. Epoch')

plt.subplot(1, 2, 2)
plt.plot(np.arange(1, len(train_losses)+1)/5, train_accuracies, label='Training Accuracy', marker='o')
plt.plot(val_accuracies, label='Validation Accuracy', marker='o')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Training and Validation Accuracy vs. Epoch')
plt.legend()

plt.tight_layout()
plt.show()
```

The SDSS and DECals labels do not exactly match for similar galaxy morphologies, hence we re-label the test images so that these fit in.

```python
cross_labels = np.copy(labels)

cross_labels[np.where(labels==2)] = 1
cross_labels[np.where(labels==3)] = 2
cross_labels[np.where(labels==4)] = 3
cross_labels[np.where(labels==8)] = 5
cross_labels[np.where(labels==9)] = 4
cross_labels[np.where(labels==6)] = 6
cross_labels[np.where(labels==7)] = 8
```

```python
cross_labels[np.where(labels==0)] = 0
cross_labels[np.where(labels==1)] = 0
cross_labels[np.where(labels==5)] = 7
```

Additionally, DECals images need to be downscaled to the same size as the SDSS images, and need to show galaxies at a similar scale. We thus cropped the images to the central 120 pixels and then downscaled it by a bilinear transform to 69 pixels on each side, which finally matched the SDSS image dataset resolution.

```python
import torch
import torch.nn.functional as F

image_tensor = torch.tensor(images, dtype=torch.float32)

# Compute coordinates for the top-left corner of the crop box
crop_size = 120
center_x = image_tensor.shape[2] // 2
center_y = image_tensor.shape[1] // 2
top_left_x = center_x - (crop_size // 2)
top_left_y = center_y - (crop_size // 2)

# Crop the images to 150x150 centered at the original images' centers
cropped_images = image_tensor[:, top_left_y:top_left_y+crop_size, top_left_x:top_left_x+crop_size, :]

# Use bilinear interpolation to downscale the cropped images to 69x69
downscaled_images = F.interpolate(cropped_images.permute(0, 3, 1, 2), size=(69, 69), mode='bilinear', align_corners=False)

# Convert the tensor back to a numpy array
downscaled_images_array = downscaled_images.permute(0, 2, 3, 1).numpy()  # Change back to NHWC format

# Ensure the values are within the valid range [0, 255]
downscaled_images_array = np.clip(downscaled_images_array, 0, 255).astype(np.uint8)
```
